A function in Python is a block of reusable code defined using def, and it can be called independently.

Example:
Function:

    def add(a, b):  
    return a + b

Method:

    class MyClass:  
      def greet(self):  
        print("Hello")  

    obj = MyClass()  
    obj.greet()  # Output: Hello

#5. What are iterators in Python and how do they differ from iterables?
Iterators vs. Iterables in Python
Iterable: An object that can be looped over (e.g., list, tuple, string). It has the __iter__() method.

Iterator: An object that remembers its state during iteration. It has both __iter__() and __next__() methods.
Example:

**Iterable (list)**

    numbers = [1, 2, 3]
    iterator = iter(numbers)  # Convert iterable to iterator

    print(next(iterator))  # Output: 1
    print(next(iterator))  # Output: 2
    print(next(iterator))  # Output: 3
 Key Difference: Iterables can be converted into iterators using iter(), while iterators generate values one at a time using next().

#6. Explain the concept of generators in Python and how they are defined.
Generators are special functions that produce values one at a time using yield, instead of returning all values at once.

Defining a Generator:

    def count_up(n):
      for i in range(1, n + 1):
         yield i  # Suspends execution and returns value

    gen = count_up(3)
    print(next(gen))  # Output: 1
    print(next(gen))  # Output: 2
    print(next(gen))  # Output: 3
Key Benefits: Memory-efficient, lazy evaluation, and useful for large datasets.

#7. What are the advantages of using generators over regular functions?

Advantages of Generators Over Regular Functions
* Memory Efficient - Generates values one at a time instead of storing them all in memory.
* Lazy Evaluation - Produces values only when needed, improving performance.
* Faster Execution - No need to create and store large data structures.
* Infinite Sequences - Can generate infinite sequences (e.g., Fibonacci, streams).
* Simpler Code - Uses yield, avoiding complex state management.
Example:

    def count_up(n):
      for i in range(1, n + 1):
        yield i  # Generates values one by one
Generators are ideal for large datasets and streaming data. ðŸš€

#9. Explain the purpose and usage of the `map()` function in Python.

map() Function in Python
The map() function applies a given function to each item in an iterable and returns a new iterable (map object).

Syntax:

    map(function, iterable)
Example:

    numbers = [1, 2, 3, 4]
    squared = list(map(lambda x: x * x, numbers))
    print(squared)  # Output: [1, 4, 9, 16]
**Purpose & Usage:**

* Applies a function to all elements in an iterable.
* Useful for transformations without explicit loops.
* Best for applying operations on lists efficiently.

#10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Difference Between map(), filter(), and reduce()

**map()** - Transforms each element using a function.


    nums = [1, 2, 3]
    squared = list(map(lambda x: x * x, nums))
    print(squared)  # Output: [1, 4, 9]

**filter()** - Selects elements that meet a condition.


    nums = [1, 2, 3, 4, 5]
    evens = list(filter(lambda x: x % 2 == 0, nums))
    print(evens)  # Output: [2, 4]
**reduce()** - Aggregates all elements into a single value.

    from functools import reduce
    nums = [1, 2, 3, 4]
    product = reduce(lambda x, y: x * y, nums)
     print(product)  # Output: 24
**map() modifies, filter() selects, and reduce() combines elements.**

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def check_prime(n):
  if n<2:
    return False
  for i in range(2, int(n**0.5)+1):
    if n%i==0:
      return False
  return True

prime=[num for num in range (1,201) if check_prime(num)]
print(prime)
```
