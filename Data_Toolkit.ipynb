#2.How does broadcasting work in NumPy?
Broadcasting allows element-wise operations between arrays of different shapes without explicitly replicating data. NumPy automatically expands smaller arrays to match the shape of larger ones, making operations faster and memory-efficient.
      
    import numpy as np
    arr = np.array([1, 2, 3])
    print(arr + 5)  
    # Output: [6 7 8]  (5 is broadcasted to [5, 5, 5])

#6.What are the differences between NumPy arrays and Python lists?
NumPy Arrays vs. Python Lists
1. Speed
NumPy arrays are faster due to C-based implementation and vectorized operations, while Python lists require loops.

2. Memory Efficiency
NumPy uses less memory because data is stored in contiguous blocks, whereas Python lists store object references.

3. Data Types
NumPy arrays are homogeneous (same type), while Python lists are heterogeneous (mixed types).

4. Mathematical Operations
NumPy supports built-in vectorized operations, whereas Python lists rely on loops.

5. Advanced Features
NumPy supports multi-dimensional arrays, broadcasting, linear algebra, and random sampling, which Python lists do not.

Key Takeaway:
Use NumPy for numerical computing (faster & efficient). Python lists are flexible but slower for large data.

#17.Explain the difference between apply() and map() in Pandas?

Both apply() and map() are used for applying functions to Pandas data, but they have key differences.

map():

Used only on Series (single column).
Works with functions, dictionaries, or lambda expressions.
Applies element-wise operations.

    import pandas as pd  
    df = pd.DataFrame({'A': [1, 2, 3, 4]})  
    df['A'] = df['A'].map(lambda x: x * 2)  
    print(df)  # Multiplies each value by 2
apply():

Works on both Series and DataFrames (rows/columns).
Can apply more complex functions.
More flexible than map().

    df['A'] = df['A'].apply(lambda x: x ** 2)  
    print(df)  # Squares each value
Key Difference:

Feature	map()	apply()

Works on	Series only	Series & DataFrame
Function Type	Element-wise	Can be applied to rows/columns
Use Case	Simple operations	Complex functions

Use map() for simple transformations and apply() for more flexibility!

```python
#4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns

data = {'A': [1, 2, 3, 4, 5],
        'B': [2, 4, 6, 8, 10],
        'C': [5, 3, 8, 1, 7]}

df = pd.DataFrame(data)

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Show plot
plt.title("Correlation Matrix Heatmap")
plt.show()
```
