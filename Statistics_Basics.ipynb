#1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales

Types of Data
--------------

Qualitative data :
----------------
Describes categories or qualities.
Example: Gender, color, brand names.

Quantitative data :
----------------
Deals with numbers and measurements.
Example: Age, height, salary.

Scales of Measurement

Nominal: Categories with no order.
Example: Eye color, type of car.

Ordinal: Categories with a meaningful order, but no fixed gaps.
Example: Rankings, satisfaction levels.

Interval: Numbers with equal intervals, but no true zero.
Example: Temperature in Celsius, IQ scores.

Ratio: Numbers with equal intervals and a true zero.
Example: Height, weight, age.

# 5. Discuss the role of random sampling in making inferences about populations.

Random Sampling & Inference
Random sampling means selecting individuals from a population by chance, so every member has an equal chance of being chosen.

Why It Matters
--------------
Helps ensure the sample is representative of the population.

Reduces bias, leading to more accurate and fair conclusions.

Allows us to make valid inferences (educated guesses) about the entire population based on the sample.

Skewness
--------
Skewness shows whether the data is symmetrical or leans to one side.
It tells us about the shape of the data distribution.

Types of Skewness:

Symmetrical (No Skew)
----------------------
Left and right sides of the graph are balanced.
Mean â‰ˆ Median â‰ˆ Mode

Positive Skew (Right Skew)
--------------------------
Tail is longer on the right side.
Mean > Median > Mode

Negative Skew (Left Skew)
--------------------------
Tail is longer on the left side.

Effect on Interpretation
--------------------------
Skewness affects which measure of central tendency is most accurate.

In skewed data, the median is often better than the mean because it's less affected by extreme values

#9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

Normal Distribution
----------------------
1.Bell-shaped curve: Symmetrical around the mean.

2.Mean = Median = Mode: All are at the center.

3.68% of data lies within 1 standard deviation from the mean.

4.Tails extend infinitely in both directions.

Empirical Rule (68-95-99.7 Rule)
---------------------------------
For a normal distribution:

1. 68% of the data falls within 1 standard deviation from the mean.

2. 95% falls within 2 standard deviations.

3. 99.7% falls within 3 standard deviations.

Example:

If the mean test score is 100 with a standard deviation of 10:

68% of scores are between 90 and 110.

95% of scores are between 80 and 120.

99.7% of scores are between 70 and 130.

# 11. Explain what a random variable is and differentiate between discrete and continuous random variables

What is a Random Variable?

A random variable is a value that represents the outcome of a random event, expressed as a number. It helps us quantify uncertain situations.

Types of Random Variables:

1. Discrete Random Variable
Takes specific, separate values (usually whole numbers).

You can count the possible outcomes.

Example: The number of students in a class, or the number of times you roll a 6 on a die.

2. Continuous Random Variable
Can take any value within a range, even decimals.

You canâ€™t count the exact number of possibilities because they are infinite.

Example: The height of a person, or the time it takes to finish a test.

Main Difference:

Discrete = countable outcomes

Continuous = infinite outcomes within a range
