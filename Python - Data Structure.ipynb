3.**What are the main differences between lists and tuples in Python?**

Mutability: Lists are mutable, meaning their elements can be changed, added, or removed. Tuples are immutable, meaning their elements cannot be modified after creation.

Syntax: Lists are created using square brackets [], while tuples use parentheses ().

Example:
lst = [1, 2, 3]  # List
tup = (1, 2, 3)  # Tuple

Performance: Tuples are faster than lists because they have a fixed size and require less memory.

Memory Usage: Lists consume more memory due to their dynamic nature, while tuples are more memory-efficient.

Functions: Lists provide more built-in functions like append(), remove(), and sort(), while tuples have limited functions like count() and index().

Use Case: Use lists when you need to modify data, and use tuples when you want data to remain unchanged.

8.**What is a hash table, and how does it relate to dictionaries in Python?**

A hash table is a data structure that stores key-value pairs using a hashing function to map keys to unique memory locations.

How It Relates to Dictionaries in Python?

Python dictionaries are implemented using hash tables internally.

Each key is hashed using the hash() function.
The computed hash value determines where the value is stored.
This makes dictionary operations (search, insert, delete) very efficient.

14.**How does the “in” keyword work differently for lists and dictionaries?**
For Lists:
The in keyword checks if an element is present in the list. It searches through the list's values.

my_list = [10, 20, 30]
print(20 in my_list)  # Output: True
print(40 in my_list)  # Output: False

For Dictionaries:
The in keyword checks if a key exists in the dictionary, not the value. It looks through the keys of the dictionary.

Example:
my_dict = {"name": "Alice", "age": 25}
print("name" in my_dict)  # Output: True
print("Alice" in my_dict)  # Output: False (Looks for keys, not values)

17.**Describe the time complexity of accessing elements in a dictionary**

The time complexity of accessing elements in a Python dictionary is O(1) on average. This means that accessing a value using a key is a constant-time operation, regardless of the size of the dictionary.

why O(1) Time Complexity?
Dictionaries in Python are implemented using hash tables:

When you access a value using a key (dict[key]), Python hashes the key to compute a unique index and retrieves the value directly from that index.
This direct access avoids the need to search through the entire dictionary, resulting in constant time complexity for lookups.

18.In what situations are lists preferred over dictionaries?

19.**Why are dictionaries considered unordered, and how does that affect data retrieval?**

Before Python 3.7, dictionaries were considered unordered because they did not guarantee the order of elements. This was due to their implementation using hash tables, which do not preserve the insertion order of items. The keys in the dictionary are hashed to determine their location, and this process doesn't maintain any specific order.

However, starting from Python 3.7, dictionaries maintain insertion order, meaning that items are stored in the order they were added. But, technically, they are still considered unordered collections because the underlying structure (hash table) doesn't inherently require order. The order is just an implementation detail in modern versions.

How This Affects Data Retrieval?

Before Python 3.7:

Data retrieval by key is still efficient (O(1) time complexity), but you cannot rely on the order of elements when iterating over the dictionary.
The dictionary might return keys in an arbitrary order.

Example:

my_dict = {"name": "Alice", "age": 30}

for key in my_dict:

    print(key)    # The order is unpredictable in versions before 3.7

20.**Explain the difference between a list and a dictionary in terms of data retrieval.**

difference Between List and Dictionary in Terms of Data Retrieval

Data Structure:
List: A list is an ordered collection that stores elements by index. Each item in the list has a unique position (index), and data is retrieved using this index.
Dictionary: A dictionary is a collection of key-value pairs. Data is retrieved using keys (not indices).

Data Retrieval:
List: Data is accessed by indexing, meaning you specify the position of the item in the list. Retrieval is done using the index value (e.g., my_list[2]).

Time Complexity: O(1) for direct access by index.
Example:

my_list = [10, 20, 30]
print(my_list[1])  # Output: 20

Dictionary: Data is accessed by key. You reference the dictionary using a specific key (e.g., my_dict["age"]).

Time Complexity: O(1) on average due to the hash table implementation.
Example:

my_dict = {"name": "Alice", "age": 30}
print(my_dict["name"])  # Output: Alice

Lookup Mechanism:

List: The data is retrieved by its index, which is sequential and starts from 0.
Dictionary: The data is retrieved by hashing the key to quickly access the corresponding value.

Order:

List: Lists are ordered collections, so elements are retrieved in the order they were added.
Dictionary: Starting from Python 3.7, dictionaries also preserve insertion order, but you still need to use keys for access.
